```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.ext.automap import automap_base
from sqlalchemy import Table, Column, Integer, String, Float, DateTime, MetaData
```

```python
engine = create_engine("sqlite:///../data.sqlite", echo=False)
conn = engine.connect()

Base = automap_base()
Base.prepare(engine, reflect=True)
meta = MetaData()
```

```python
tweets_data = pd.read_sql('SELECT * FROM TWEETS', conn)
tweets_data = tweets_data.rename(columns={"created_at":"date"})
date_time = pd.DataFrame(tweets_data.date.str.split(' ',1).tolist(),columns = ['date','time'])
tweets_data["date"] = date_time["date"]
tweets_data["time"] = date_time["time"]
tweets_data.head()
```

```python
emp_matchers = ["employ", "jobs"]
emp_matching = [s for s in tweets_data["text"] if any(xs in s for xs in emp_matchers)]
```

```python
employ_df = tweets_data[tweets_data["text"].isin(emp_matching)]
employ_df.head()
```

```python
employ_df.to_csv("employment_tweets.csv")
```
